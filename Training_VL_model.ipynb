```python
from pycaret.classification import ClassificationExperiment
import pandas as pd
```

```python
season_data = pd.read_csv('my_data_02.csv')
```

```python
len(season_data)
```

```python
clf = ClassificationExperiment()
clf.setup(data=season_data, target='Result', categorical_features=['Home Team', 'Away Team', 'Season'],
          normalize=True, normalize_method='minmax', transformation=True, transformation_method='yeo-johnson',
          session_id=42, use_gpu=True, remove_outliers=True)
```

```python
best_model = clf.compare_models(sort='Accuracy', turbo=True)
```

```python
clf.save_model(best_model, 'my_best_model_LGBM')
```
